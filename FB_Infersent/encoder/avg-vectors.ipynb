# Conjunction split

```python
# Load sentences
sentences = []
indi_sent_rev = []
with open('text_0_AutomotiveProd7.txt') as f:
    i = 1
    rev_num = -1
    for line in f:
        if i == 1:
            rev_num += 1
            l = line.strip()
            indi_sent = l.split('.')
            indi_sent = [a.strip() for a in indi_sent]
            indi_sent = list(filter(None, indi_sent))
            for k in range(len(indi_sent)):
                indi_sent_rev.append(rev_num)
            sentences.extend(indi_sent)
            i *= -1
        else:
            i *= -1

print(len(sentences))
```

```python
new_sentences = []
new_indi_sent_rev = []
for ind, sent in enumerate(sentences):
    if ' and ' in sent:
        temp = sent.split(' and ')
        temp_rev = [indi_sent_rev[ind] for i in range(len(temp))]
        new_sentences.extend(temp)
        new_indi_sent_rev.extend(temp_rev)
    elif ' or ' in sent:
        temp = sent.split(' or ')
        temp_rev = [indi_sent_rev[ind]] * len(temp)
        new_sentences.extend(temp)
        new_indi_sent_rev.extend(temp_rev)
    else:
        new_sentences.append(sent)
        new_indi_sent_rev.append(indi_sent_rev[ind])
```

```python
new_words = []
for i in new_sentences:
    new_words.append(i.split(' '))
```

```python
print(len(new_words))
```

# Embeddings

```python
model = gensim.models.Word2Vec(new_words, size=300, window=5, min_count=1)
```

```python
model.save('word2vec_model2')
```

```python
model.wv['much']
```

```python
embeddings = []
for sent in new_sentences:
    temp_embeddings = model[sent.split(' ')[0]]
    for word in sent.split(' ')[1:]:
        temp_embeddings = np.add(temp_embeddings, model[word])
    temp_embeddings = temp_embeddings / len(sent.split(' '))
    embeddings.append(temp_embeddings)
```

# Split clusters

```python
from sklearn.cluster import SpectralClustering
from sklearn import metrics

# Spectral Clustering
sc = SpectralClustering(30, affinity='nearest_neighbors', n_init=1000)
sc.fit(embeddings)
print('spectral clustering')
print(sc.labels_)
```

```python
labels = sc.labels_.flatten()
print(labels.shape)
labels = list(labels)
for i in set(sc.labels_):
    print(str(i)+': '+str(labels.count(i)))
```

```python
cluster_new = []
for i in range(30):
    temp_sentences = []
    pos_sent = []
    neg_sent = []
    #print('Cluster '+str(i)+':')
    for k, l in enumerate(new_sentences):
        if sc.labels_[k] == i:
            temp_sentences.append(k)
    for sent in temp_sentences:
        if ratings[new_indi_sent_rev[sent]] >= 3.0:
            pos_sent.append(sent)
        else:
            neg_sent.append(sent)
    cluster_new.append(pos_sent)
    cluster_new.append(neg_sent)
```

```python
matrix = []
for i in range(rev_num+1):
    matrix_row = []
    for clusters in cluster_new:
        flag = 0
        for sent_ind in clusters:
            if new_indi_sent_rev[sent_ind] == i:
                matrix_row.append('1')
                flag = 1
                break
        if flag == 0:
            matrix_row.append('0')
    matrix.append(matrix_row)
```
